```python
transactions=df
```

```python
transactions['transaction_date'] = pd.to_datetime(transactions['Date']).dt.date
transactions['transaction_time'] = pd.to_datetime(transactions['Date']).dt.time
```

```python

def determine_shift(transaction_time):
    if transaction_time >= pd.to_datetime("09:00:00").time() and transaction_time <= pd.to_datetime("17:00:00").time():
        return 1
    elif transaction_time > pd.to_datetime("17:00:00").time() or transaction_time <= pd.to_datetime("00:00:00").time():
        return 2
    elif transaction_time >= pd.to_datetime("00:01:00").time() and transaction_time < pd.to_datetime("09:00:00").time():
        return 3
    else:
        return "Unknown"
transactions['shift'] = transactions['transaction_time'].apply(determine_shift)
```
